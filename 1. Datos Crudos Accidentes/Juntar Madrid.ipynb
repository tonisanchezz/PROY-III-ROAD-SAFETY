```python
import pandas as pd

# Pasamos del formato antiguo (Mayúsculas/Espacios) al formato nuevo (minúsculas/guiones_bajos)
mapeo_columnas = {
    'FECHA': 'fecha',
    'RANGO HORARIO': 'hora',
    'DISTRITO': 'distrito',
    'DIA SEMANA': 'dia_semana', 
    'LUGAR ACCIDENTE': 'localizacion',
    'Nº': 'numero',
    'Nº PARTE': 'num_expediente',
    'TIPO ACCIDENTE': 'tipo_accidente',
    'Tipo Vehiculo': 'tipo_vehiculo',
    'TIPO PERSONA': 'tipo_persona',
    'SEXO': 'sexo',
    'LESIVIDAD': 'lesividad',
    'Tramo Edad': 'rango_edad',
    # Unificamos las 3 variantes de víctimas
    'Nº VICTIMAS * ': 'num_victimas',
    'Nº VICTIMAS *': 'num_victimas',
    '* Nº VICTIMAS': 'num_victimas'
}

dfs_listos = []

for i in range(2016, 2024):
    archivo = f"madrid {i}.xlsx"
    df = pd.read_excel(archivo)
    df = df.rename(columns=mapeo_columnas)        
    dfs_listos.append(df)
    print(f"Procesado {archivo}: {df.shape[1]} columnas")

df_final = pd.concat(dfs_listos, ignore_index=True, join="outer")

print(f"\nTotal filas final: {len(df_final)}")
print(f"Total columnas final: {len(df_final.columns)}")
print(f"Columnas resultantes: {df_final.columns.tolist()}")
```
