# Variables
## ID
* **num_expediente**: ID del accidente (juntado num_expediente y PARTE)  
Entero, 0% faltantes
## Tiempo
* **fecha**: dd/mm/aaaa  
Date, 0% faltantes
* **hora**: formato 24h (hay que arreglar el formato de 2016-2018)  
text/time, 0% faltantes
* **dia_semana**: LUNES, MARTES...  
categorico, 71% faltantes
## Espacio
* **distrito**: distrito de madrid  
categorico, <0.1% faltantes
* **cod_distrito**: distrito de madrid (código)  
entero, 28% faltantes
* **localizacion**: carretera/avenida/calle/... del accidente  
texto, 0% faltantes
* **numero**: número de la calle (si pone 0 en 2016-2018 creo que no lo podemos localizar exactamente al no tener tampoco coordenadas)  
entero/texto, <0.1% faltantes
* **coordenada_x_utm**: coordenada x en utm (solo 2019-2024)  
float, 28.7% faltantes
* **coordenada_y_utm**: coordenada y en utm (solo 2019-2024)  
float, 28.7% faltantes
## Accidente
* **tipo_accidente**: tipo de accidente (atropello, colisión, alcance...)  
categorica, <0.1% faltantes
* **num_victimas**: Número de víctimas en el accidente (puede que los faltantes sean que no hayan)  
entero, 71.3% faltantes
* **tipo_vehiculo**: tipo vehiculo (turismo, motocicleta...). "No asignado" para peatones  
text/categorica, <0.1% faltantes
## Entorno
Estado y entorno de la carretera  
CPFA: Condiciones ambientales que se dan en el momento del siniestro  
CPSV: Circunstancias que afectan al estado de la vía en el momento de producirse el
siniestro
* **CPFA Granizo**: SI o NO (solo 2016-2018)  
binaria, 71.3% faltantes
* **CPFA Hielo**: SI o NO (solo 2016-2018)  
binaria, 71.3% faltantes
* **CPFA Lluvia**: SI o NO (solo 2016-2018)  
binaria, 71.3% faltantes
* **CPFA Niebla**: SI o NO (solo 2016-2018)  
binaria, 71.3% faltantes
* **CPFA Seco**: SI o NO (solo 2016-2018)  
binaria, 71.3% faltantes
* **CPFA Nieve**: SI o NO (solo 2016-2018)  
binaria, 71.3% faltantes
* **CPSV Mojada**: SI o NO (solo 2016-2018)  
binaria, 71.3% faltantes
* **CPSV Aceite**: SI o NO (solo 2016-2018)  
binaria, 71.3% faltantes
* **CPSV Barro**: SI o NO (solo 2016-2018)  
binaria, 71.3% faltantes
* **CPSV Grava Suelta**: SI o NO (solo 2016-2018)  
binaria, 71.3% faltantes
* **CPSV Hielo**: SI o NO (solo 2016-2018)  
binaria, 71.3% faltantes
* **CPSV Seca y Limpia**: SI o NO (solo 2016-2018)  
binaria, 71.3% faltantes
* **estado_meteorológico**: Despejado, lluvía débil... (solo 2019-2024)  
categórica, 36.3% faltantes
## Persona
* **tipo_persona**: conductor, pasajero, peatón, testigo...  
categorica, <0.1% faltantes
* **sexo**: Hombre, mujer o desconocido.  
categorica, 0% faltante
* **lesividad**: Ingresado, heridas leves, fallecido... (faltantes sin asistencia sanitaria)  
categorica, 32.3% faltante
* **cod_lesividad**: Codificación de la lesivilidad (solo en 2019-2024)  
entero, 61% faltantes
* **rango_edad**: rango de edad de la persona  
categorica, 0% faltante

## Pruebas
* **positiva_alcohol**: S o N. La gran mayoría dan negativo (los faltantes es que no le hacen la prueba)  
categorica/binaria, 28.9% faltantes
* **positiva_droga**: NULL o 1. NULL serán los que dan negativo (o no le hacen la prueba)  
categorica. 99.8% faltantes
