```python
import pandas as pd
print("Procesando festividades...")
df_l = pd.read_csv('festivos_locales_historicos.csv', sep=';', encoding='latin-1')
df_r = pd.read_csv('festivos_regionales_historicos.csv', sep=';', encoding='latin-1')
df_acc = pd.read_csv('madrid_accidentes_limpio.csv', encoding='latin-1')

# Filtrar años 2016-2024 y municipio Madrid
h_dates = set(df_r[df_r['año'].between(2016, 2024)]['fecha_festivo'].astype(str))
h_dates.update(df_l[(df_l['año'].between(2016, 2024)) & (df_l['municipio_nombre'].str.contains('Madrid', na=False))]['fecha_festivo'].astype(str))

df_acc['es_festivo'] = df_acc['fecha'].astype(str).isin(h_dates).astype(int)
df_acc.to_csv('madrid_accidentes_limpio_festivos.csv', index=False)

print("Proceso finalizado.")
```
